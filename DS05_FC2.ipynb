```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

```python
np.random.seed(0)
ad_spend = np.linspace(10,100,24 )# in thousands$
print(ad_spend)
```

```python
sales_return = 50 + 3.5 * ad_spend + np.random.normal(0,20,24 )
print(sales_return)
```

```python
X_ad_spend = sm.add_constant(ad_spend)
model = sm.OLS(sales_return, X_ad_spend)
results_fit = model.fit()
```

```python
predicted_sales = results_fit.predict(X_ad_spend)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(ad_spend, sales_return ,color = 'green')
plt.plot(ad_spend, predicted_sales, color = 'red')
plt.show()
results_fit.summary()
```

0.956 - r squared - variance that can be explained in the model which is good fit




MULTIPLE REGRESSION
1. y = b0  + B1X1 + B2X2 + B3X3 + B4X4 + BnXn

y = sales of handbags
x1 = advertising budget spend
x2 = no of celeb endorsement
x3 = review score
x4 = proximity

```python
np.random.uniform(3,5,100)
```

```python
#simulate x1, x2,x3, x4
ad_spend =  np.linspace(50,500,100)# amount spend in thousands$
celebs_endosr = np.random.randint(1,5,100) # number of celebs from 1-5
cust_rating= np.random.uniform(3,5,100) # rating between 3-5
proximity = np.random.randint(5,20,100) # 5min to 20min walk
```

```python
sales_return = ( 50 +
                0.5 * ad_spend  +
                20 * celebs_endosr +
                10 * cust_rating  +
                0.5 * proximity
                + np.random.normal(0,20,100 ))
print(sales_return)
```

```python
df_gucci = pd.DataFrame(
    {
        'Sales': sales_return,
        'Ad Spend':  ad_spend,
        'Celebrity Endorsements': celebs_endosr,
        'Cstomer rating': cust_rating ,
        'Near to center' : proximity

    })
```

```python
Y= df_gucci.Sales
X = df_gucci[['Sales','Ad Spend', 'Celebrity Endorsements', 'Cstomer rating', 'Near to center']]

X = sm.add_constant(X)
model = sm.OLS(Y,X)
results_fit = model.fit()
```

```python
predicted_sales_multi = results_fit.predict(X)
print(predicted_sales_multi)
print(sales_return, predicted_sales_multi)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(celebs_endosr, sales_return ,color = 'green')
plt.plot(celebs_endosr, predicted_sales_multi, color = 'red')
plt.show()
results_fit.summary()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(celebs_endosr, sales_return ,color = 'green')
plt.plot(celebs_endosr, predicted_sales_multi, color = 'red')
plt.show()
results_fit.summary()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(cust_rating, sales_return ,color = 'green')
plt.plot(cust_rating, predicted_sales_multi, color = 'red')
plt.show()
results_fit.summary()
```

```python
plt.figure(figsize = (10,6))
plt.scatter(proximity, sales_return ,color = 'green')
plt.plot(proximity, predicted_sales_multi, color = 'red')
plt.show()
results_fit.summary()
```
